```python
import os
files = os.listdir('/lakehouse/default/Files/REFERENCE')
for f in files:
    print(f)
```

```python
import pandas as pd
pdf = pd.read_excel(
    '/lakehouse/default/Files/REFERENCE/imd_la_summary.xlsx',
  sheet_name='IMD',
    header=0
)

print(f'Rows: {len(pdf)}')
print(f'Columns: {list(pdf.columns)}')
pdf.head(5)
```

```python
pdf.columns = (pdf.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('-', '_')
    .str.replace('(', '')
    .str.replace(')', '')
    .str.replace('%', 'pct')
    .str.replace('/', '_')
)

print('Cleaned column names:')
for c in pdf.columns:
    print(f'  {c}')
```

```python
df = spark.createDataFrame(pdf)
df.write.format('delta').mode('overwrite').saveAsTable('bronze_imd')
print('IMD table created successfully')
```
