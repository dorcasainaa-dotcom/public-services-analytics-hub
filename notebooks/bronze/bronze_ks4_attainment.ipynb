# Bronze: KS4 Attainment
**Source:** MHCLG LA Revenue Account Outturn
**Purpose:** Load raw CSV into Bronze Delta table
**Date loaded:** 7th April 2026
**Rows expected:** approximately 300 (one per local authority)

```python
df = spark.read.csv('Files/ks4_attainment.csv', header=True,
inferSchema=True)
print(f'Rows: {df.count()}')
print(f'Columns: {len(df.columns)}')
df.printSchema()
```

```python
df.show(5, truncate=False)
```

```python
from pyspark.sql.functions import col, count, when
null_counts = df.select([count(when(col(c).isNull(), c)).alias(c) for c in
df.columns])
null_counts.show()
```

```python
df.write.format('delta').mode('overwrite').saveAsTable('bronze_ks4_attainment')
```
